```python
#Apply salt and pepper noise and remove the noise using min and max filtering technique. Show input and output side by side.
img = cv2.imread('/content/Picture1.jpg',cv2.IMREAD_GRAYSCALE)

def add_salt_pepper_noise(image, salt_prob=0.05, pepper_prob=0.05):
    noisy = image.copy()
    total_pixels = image.size

    # Salt noise (white pixels)
    num_salt = int(salt_prob * total_pixels)
    coords = [np.random.randint(0, i - 1, num_salt) for i in image.shape]
    noisy[coords[0], coords[1]] = 255

    # Pepper noise (black pixels)
    num_pepper = int(pepper_prob * total_pixels)
    coords = [np.random.randint(0, i - 1, num_pepper) for i in image.shape]
    noisy[coords[0], coords[1]] = 0

    return noisy

noisy_img = add_salt_pepper_noise(img)

#  Apply Min Filter
min_filtered = cv2.erode(noisy_img, np.ones((3, 3), np.uint8))

#  Apply Max Filter
max_filtered = cv2.dilate(noisy_img, np.ones((3, 3), np.uint8))

#  Show Results Side by Side
titles = ['Original', 'Salt & Pepper Noise', 'Min Filter', 'Max Filter']
images = [img, noisy_img, min_filtered, max_filtered]

plt.figure(figsize=(12, 6))
for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()


```

```python
#Apply Gaussian noise to the following image and remove the noise using Gaussian filtering. Show input and output side by side.
img = cv2.imread('/content/Picture1.jpg',cv2.IMREAD_GRAYSCALE)

#  Add Gaussian noise
def add_gaussian_noise(image, mean=0, std=20):
    gaussian_noise = np.random.normal(mean, std, image.shape).astype(np.float32)
    noisy_img = image.astype(np.float32) + gaussian_noise
    noisy_img = np.clip(noisy_img, 0, 255).astype(np.uint8)
    return noisy_img

noisy_img = add_gaussian_noise(img)

# Apply Gaussian Filtering (Smoothing)
gaussian_filtered = cv2.GaussianBlur(noisy_img, (5, 5), sigmaX=1)

#  Show Original, Noisy, and Denoised Images
titles = ['Original', 'Gaussian Noise', 'Gaussian Filtered']
images = [img, noisy_img, gaussian_filtered]

plt.figure(figsize=(12, 4))
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
img = cv2.imread('/content/Picture3.jpg',cv2.IMREAD_GRAYSCALE)

# Add Gaussian noise
def add_gaussian_noise(image, mean=0, std=20):
    gaussian_noise = np.random.normal(mean, std, image.shape).astype(np.float32)
    noisy_img = image.astype(np.float32) + gaussian_noise
    noisy_img = np.clip(noisy_img, 0, 255).astype(np.uint8)
    return noisy_img

noisy_img = add_gaussian_noise(img)

# Apply Box Filtering
box_filtered = cv2.boxFilter(noisy_img, ddepth=-1, ksize=(3, 3))

# Apply Average Filtering
average_filtered = cv2.blur(noisy_img, (3, 3))

# Apply Median Filtering
median_filtered = cv2.medianBlur(noisy_img, 3)

# Display All Results (without Min Filter)
titles = ['Original', 'Gaussian Noisy', 'Box Filter', 'Average Filter', 'Median Filter']
images = [img, noisy_img, box_filtered, average_filtered, median_filtered]

plt.figure(figsize=(15, 5))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# a) Read and show the image
img = cv2.imread('/content/Picture4.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.title("Figure 3.4")
plt.axis('on')
plt.show()

# b) Show the matrix form of the image
print("b) Matrix Form of Image:\n", img)

# c) Show pixel info on hover (interactive)
# NOTE: This must be run in an interactive Python environment (like Jupyter Notebook)
def format_coord(x, y):
    col = int(x + 0.5)
    row = int(y + 0.5)
    if 0 <= col < img.shape[1] and 0 <= row < img.shape[0]:
        z = img[row, col]
        return f"x={col}, y={row}, value={z}"
    else:
        return f"x={col}, y={row}"

fig, ax = plt.subplots()
ax.imshow(img, cmap='gray')
ax.format_coord = format_coord
plt.title("c) Hover to See Pixel Info")
plt.show()

# d) Value of pixel at (10, 78)
value = img[10, 78]
print(f"d) Value of pixel (10, 78): {value}")

# e) Size of the image
print(f"e) Size (Height, Width): {img.shape}")

# f) All image information
print("\nf) Full Image Info:")
print(f" - Dimensions: {img.ndim}")
print(f" - Shape: {img.shape}")
print(f" - Total Pixels: {img.size}")
print(f" - Data Type: {img.dtype}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# a) Read and show all three types of images
rgb_img = cv2.imread("/content/Picture5.png")  # RGB image (BGR in OpenCV)
gray_img = cv2.imread("/content/Picture6.jpg", cv2.IMREAD_GRAYSCALE)  # Grayscale
indexed_img = cv2.imread("/content/Picture7.png", cv2.IMREAD_UNCHANGED)  # Indexed (simulated)

# Display images
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(rgb_img, cv2.COLOR_BGR2RGB))
plt.title("a) RGB Image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(gray_img, cmap='gray')
plt.title("a) Grayscale Image")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(indexed_img)
plt.title("a) Indexed Image")
plt.axis('off')
plt.show()

# b) Turn RGB image to Grayscale
rgb_to_gray = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2GRAY)
plt.imshow(rgb_to_gray, cmap='gray')
plt.title("b) RGB to Grayscale")
plt.axis('off')
plt.show()

# c) Indexed image to Grayscale (assume BGR)
indexed_to_gray = cv2.cvtColor(indexed_img, cv2.COLOR_BGR2GRAY)
plt.imshow(indexed_to_gray, cmap='gray')
plt.title("c) Indexed to Grayscale")
plt.axis('off')
plt.show()

# d) Indexed to RGB
indexed_to_rgb = cv2.cvtColor(indexed_img, cv2.COLOR_BGR2RGB)
plt.imshow(indexed_to_rgb)
plt.title("d) Indexed to RGB")
plt.axis('off')
plt.show()

# e) Convert Grayscale to Binary (Otsu)
_, binary_img = cv2.threshold(gray_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(binary_img, cmap='gray')
plt.title("e) Grayscale to Binary")
plt.axis('off')
plt.show()

# f) Invert the Binary image
inverted_binary = cv2.bitwise_not(binary_img)
plt.imshow(inverted_binary, cmap='gray')
plt.title("f) Inverted Binary")
plt.axis('off')
plt.show()

# g) Histogram of the Grayscale image
plt.hist(gray_img.ravel(), bins=256, range=(0, 256), color='skyblue')
plt.title("g) Histogram of Grayscale")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()

# h) Invert the RGB image
inverted_rgb = cv2.bitwise_not(rgb_img)
plt.imshow(cv2.cvtColor(inverted_rgb, cv2.COLOR_BGR2RGB))
plt.title("h) Inverted RGB")
plt.axis('off')
plt.show()

# i) Blur the RGB image
blurred_rgb = cv2.GaussianBlur(rgb_img, (7, 7), sigmaX=1.5)
plt.imshow(cv2.cvtColor(blurred_rgb, cv2.COLOR_BGR2RGB))
plt.title("i) Blurred RGB")
plt.axis('off')
plt.show()
```

Which filtering method (Box, Average, Median, Gaussian) performed best for each
noise type? Explain why.

Ans : For every noise Median is best.

Salt and Pepper Noise:
The Median Filter works best. It removes the black and white dots (salt and pepper) by picking the middle value from nearby pixels, so the extreme noisy values are ignored.

Gaussian Noise:
The Gaussian Filter is the best here. It smooths the image gently, reducing the random noise without too much blurring.
Average and Box Filters also reduce Gaussian noise, but they blur edges more.
Median Filter helps a little but not as good for this type of noise.


How does an Indexed image differ from an RGB or Grayscale image?

Ans : Grayscale images show only shades of gray  from black (0) to white (255). Each pixel has one value, and the image is stored as a 2D array.
RGB images show full color using three values per pixel: Red, Green, and Blue (each from 0 to 255). These are stored in a 3D array  height, width, and color channels.
Indexed images use a color palette. Each pixel holds just an index number, and that index points to a color in a separate colormap (which holds RGB values). This makes the file smaller, especially for images with few colors. These images are also 2D arrays, but you need the color map to see the real colors.
